```python

def lw(cols):
    Pclass=cols[1]
    age=cols[0]

    if pd.isnull(age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 30
        elif Pclass==3:
            return 25
        else:
            return 30
    else:
        return age
```

```python
age=ds[['Age', 'Pclass']].apply(lw,axis=1)
```

```python
x['Age']=age
```

```python
sb.heatmap(x.isnull(), cmap='viridis')
```
